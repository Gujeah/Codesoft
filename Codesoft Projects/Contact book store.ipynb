```python
import tkinter as tk
from tkinter import messagebox

class ContactBookApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Contact Book")

        self.name_label = tk.Label(master, text="Name:")
        self.name_label.grid(row=0, column=0, sticky=tk.E)
        self.name_entry = tk.Entry(master)
        self.name_entry.grid(row=0, column=1)

        self.phone_label = tk.Label(master, text="Phone Number:")
        self.phone_label.grid(row=1, column=0, sticky=tk.E)
        self.phone_entry = tk.Entry(master)
        self.phone_entry.grid(row=1, column=1)

        self.email_label = tk.Label(master, text="Email:")
        self.email_label.grid(row=2, column=0, sticky=tk.E)
        self.email_entry = tk.Entry(master)
        self.email_entry.grid(row=2, column=1)

        self.add_button = tk.Button(master, text="Add Contact", command=self.add_contact)
        self.add_button.grid(row=3, column=0, columnspan=2, pady=10)

        self.search_button = tk.Button(master, text="Search Contact", command=self.search_contact)
        self.search_button.grid(row=4, column=0, columnspan=2, pady=10)

        self.delete_button = tk.Button(master, text="Delete Contact", command=self.delete_contact)
        self.delete_button.grid(row=5, column=0, columnspan=2, pady=10)

        self.update_button = tk.Button(master, text="Update Contact", command=self.update_contact)
        self.update_button.grid(row=6, column=0, columnspan=2, pady=10)

        self.contact_book = ContactBook()

    def add_contact(self):
        name = self.name_entry.get()
        phone_number = self.phone_entry.get()
        email = self.email_entry.get()
        if name and phone_number and email:
            self.contact_book.add_contact(name, phone_number, email)
            messagebox.showinfo("Success", "Contact added successfully.")
        else:
            messagebox.showerror("Error", "Please fill in all fields.")

    def search_contact(self):
        phone_number = self.phone_entry.get()
        if phone_number:
            index = self.contact_book.search_contact(phone_number)
            if index is not None:
                messagebox.showinfo("Contact Found", f"Contact found at index {index}.")
            else:
                messagebox.showinfo("Contact Not Found", "Contact not found.")
        else:
            messagebox.showerror("Error", "Please enter a phone number to search.")

    def delete_contact(self):
        phone_number = self.phone_entry.get()
        if phone_number:
            self.contact_book.delete_contact(phone_number)
            messagebox.showinfo("Success", "Contact deleted successfully.")
        else:
            messagebox.showerror("Error", "Please enter a phone number to delete.")

    def update_contact(self):
        name = self.name_entry.get()
        phone_number = self.phone_entry.get()
        email = self.email_entry.get()
        if name and phone_number:
            self.contact_book.update_contact(phone_number, name, email)
            messagebox.showinfo("Success", "Contact updated successfully.")
        else:
            messagebox.showerror("Error", "Please enter both name and phone number to update.")

class ContactBook:
    def __init__(self):
        self.contacts = []

    def add_contact(self, name, phone_number, email):
        contact = Contact(name, phone_number, email)
        self.contacts.append(contact)

    def search_contact(self, phone_number):
        for i, contact in enumerate(self.contacts):
            if contact.phone_number == phone_number:
                return i
        return None

    def delete_contact(self, phone_number):
        index = self.search_contact(phone_number)
        if index is not None:
            del self.contacts[index]

    def update_contact(self, phone_number, new_name=None, new_email=None):
        index = self.search_contact(phone_number)
        if index is not None:
            if new_name:
                self.contacts[index].name = new_name
            if new_email:
                self.contacts[index].email = new_email

class Contact:
    def __init__(self, name, phone_number, email):
        self.name = name
        self.phone_number = phone_number
        self.email = email

def main():
    root = tk.Tk()
    app = ContactBookApp(root)
    root.mainloop()

if __name__ == "__main__":
    main()

```
